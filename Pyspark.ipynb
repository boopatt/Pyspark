```python
from pyspark import SparkContext, SparkConf
```

```python
# Initializing Spark
conf= SparkConf().setAppName("testApp1").setMaster("local[*]")
sc=SparkContext(conf=conf)
```

```python
print(sc)
```

```python
sc.defaultParallelism
```

```python
import random
randomlist=random.sample(range(0, 40),10)
print(randomlist)
```

```python
rdd1=sc.parallelize(randomlist,4)
rdd1.collect()
```

```python
print(rdd1.getNumPartitions())
print(rdd1.glom().collect())
print("The Two Partitions :",rdd1.glom().take(2))
```

```python
rdd1.glom().collect()[3]
```

```python
Type:1
def myfunc(item):
  return(item+1)*3
```

```python
rdd_map=rdd1.map(myfunc)
rdd_map.collect()
```

```python
Type:2
rdd_map=rdd1.map(lambda item:(item+1)*3)
rdd_map.collect()
```

```python
rdd_filter=rdd1.filter(lambda x:x%3==0)
rdd_filter.collect()
```

```python
rdd_flatmap=rdd1.map(lambda x:[x+2,x+5])

rdd_flatmap.glom().collect()
```

```python
rdd_flatmap=rdd1.flatMap(lambda X: [X+2,X+5])
rdd_flatmap.collect()
rdd_flatmap.glom().collect()
```

```python
rdd_flatmap=rdd1.flatMap(lambda X: [X+2,X+5])
rdd_flatmap.reduce(lambda x,y:x+y)
```

```python
print([rdd1.max(),rdd1.min(),rdd1.mean(),round(rdd1.stdev()),rdd1.sum()])
```

```python
def myfunc(partition):
  sum=0
  for item in partition:
    sum=sum+item
  yield sum

rdd1.mapPartitions(myfunc).collect()
```
